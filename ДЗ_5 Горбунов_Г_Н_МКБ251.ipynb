## Имеет инициализатор и 4 метода:
### Инициализатор __init__
1) Если не получилось создать счет (например, введена отрицательная сумма при инициализации объекта) - выбрасываем исключение, а обрабатывать будем в GorbunoffBank'е. Не будет обработки - прервется, будет - что-нибудь сообщим пользователю и продолжим.
2) Заполняем атрибуты начальными значениями и (я так захотел: записываем в журнал время создания счета).

### _log_operation
Приватные методы описываются первыми - поэтому следующим методом идет "логгер". "Логгер" принимает тип операции (снятие\сополнение\ создание), сумму операции, ее статус (успешно\отклонена) и баланс после выполнения операции.
Создает словарь и добавляет его в список.

### deposit()
Принимает сумму и если отрицательная - логирует ошибку, если положительная - плюсует к балансу и, опять же, логгирует.

### Withdrow()
Принимает сумму и сравнивает с балансом. Если денег хватает - снимает и логгирует, нет - логгирует ошибку, баланс неизменен.

### get_balance()
Просто возвращает float.

### get_history()
Возвращает список словарей в сыром виде Никаких приведений в строки и тому подоюное. Обрабатывать и забирать нужные поля должен "заказчик", я считаю.  

```python
# Account.py
from datetime import datetime


class Account:
    """
    Класс, моделирующий банковский счёт с ведением истории операций
    и базовыми операциями.
    """

    def __init__(self, account_holder: str, balance: float = 0.0):
        """
        Инициализирует банковский счёт.

        :param account_holder: Имя владельца счёта.
        :param balance: Начальный баланс (по умолчанию 0.0). Неотрицательный
        :raises ValueError: Если начальный баланс отрицательный.
        """
        if balance < 0:
            raise ValueError("Начальный баланс не может быть отрицательным.")
        self.holder = account_holder
        self._balance = balance
        self.operations_history = []

        # Логируем создание счёта как первую запись в истории
        self._log_operation("open_account", 0.0, "success", self._balance)

    def _log_operation(
        self, operation_type,
        amount, status, balance_after
        ) -> bool:
        """
        Внутренний метод для добавления записи об операции в историю.

        :param operation_type: Тип операции (например, 'deposit', 'withdraw').
        :param amount: Сумма операции.
        :param status: Статус операции ('success' или 'fail').
        :param balance_after: Баланс после операции.
        """
        self.operations_history.append({
            "type": operation_type,
            "amount": amount,
            "timestamp": datetime.now(),
            "balance_after": balance_after,
            "status": status
        })

    def deposit(self, amount) -> bool:
        """
        Пополняет счёт на указанную сумму.

        :param amount: Сумма пополнения (должна быть положительной).
        :return: True, если операция успешна; иначе False.
        """
        if amount <= 0:
            self._log_operation("deposit", amount, "fail", self._balance)
            return False
        self._balance += amount
        self._log_operation("deposit", amount, "success", self._balance)
        return True

    def withdraw(self, amount) -> bool:
        """
        Снимает указанную сумму со счёта.

        :param amount: Сумма снятия (должна быть положительной).
        :return: True, если операция успешна; иначе False.
        """
        if amount <= 0:
            self._log_operation("withdraw", amount, "fail", self._balance)
            return False
        if amount > self._balance:
            self._log_operation("withdraw", amount, "fail", self._balance)
            return False
        self._balance -= amount
        self._log_operation("withdraw", amount, "success", self._balance)
        return True

    def get_balance(self) -> float:
        """
        Возвращает текущий баланс счёта.

        :return: Текущий баланс.
        """
        return self._balance

    def get_history(self) -> list:
        """
        Возвращает полную историю операций.
        :return: Список записей об операциях.
        """
        return self.operations_history
```

## Задание № 2 (Класс CreditAccount)

Практически ничем не отличается, разве что:

не может быть иницирован с отрицательным балансом, а кредитный должен это позволять.
2) Логгер высчитыват остаток кредитных средств. Я посчитал, что заводить новые атрибуты класса, объявленные в нем, нецелесообразно, т.к кроме логгера ими никто не пользуется.

4) deposit() не переопределял

Добавлены:

1) get_credit_available() - вернет значение, которе еще можем потратить из кредита (по формуле из задания)
1) get_credit_limit() - вернет лимит кредита

```python
# CreditAccount.py
from datetime import datetime


class CreditAssistantMixin:
    """Вспомогательный миксин для общих методов (не обязателен, но можно)"""
    pass


class CreditAccount(Account):
    """
    Кредитный счёт, расширяющий базовый банковский счёт возможностью ухода
    в минус в пределах кредитного лимита.
    """

    def __init__(
        self, account_holder: str,
        balance: float = 0.0,
        credit_limit: float = 0.0
    ):
        """
        Инициализирует кредитный счёт.

        :param account_holder: Имя владельца счёта.
        :param balance: Начальный баланс (может быть отрицательным,
                        но не ниже -credit_limit).
        :param credit_limit: Кредитный лимит (должен быть >= 0).
        :raises ValueError: Если credit_limit < 0 или balance < -credit_limit.
        """
        # Сначала проводим свою валидацию
        if credit_limit < 0:
            raise ValueError("Кредитный лимит не может быть отрицательным.")
        if balance < -credit_limit:
            raise ValueError("Начальный баланс не может быть ниже кред. лимита")

        # Вызываем родительский __init__ с временным безопасным балансом
        super().__init__(account_holder, 0.0)

        # Теперь корректируем баланс и добавляем свои атрибуты
        self._balance = balance
        self._credit_limit = credit_limit

        # Перезаписываем историю: удаляем запись "open_account" от родителя
        # и логируем свою
        self.operations_history.clear()
        self._log_operation(
            operation_type="open_account",
            amount=0.0,
            status="success",
            balance_after=self._balance
        )

    def _log_operation(self, operation_type, amount, status, balance_after):
        """
        Логирует операцию с полной информацией для последующего форматирования.

        :param operation_type: Тип операции (например, 'deposit', 'withdraw').
        :param amount: Сумма операции.
        :param status: Статус операции: 'success' или 'fail'.
        :param balance_after: Фактический баланс счёта после выполнения операции.
        """
        own_funds = max(0.0, balance_after)
        credit_used = max(0.0, -balance_after)
        credit_available = max(0.0, self._credit_limit - credit_used)

        self.operations_history.append({
            "type": operation_type,
            "amount": amount,
            "timestamp": datetime.now(),
            "balance_after": balance_after,
            "status": status,
            "credit_limit": self._credit_limit,
            "own_funds_after": own_funds,
            "credit_available_after": credit_available
        })

    def withdraw(self, amount: float) -> bool:
        """
        Снимает средства. Разрешено уходить в минус в пределах кред. лимита.

        :param amount: Сумма снятия (должна быть > 0).
        :return: True при успехе, иначе False.
        """
        if amount <= 0:
            self._log_operation("withdraw", amount, "fail", self._balance)
            return False

        new_balance = self._balance - amount
        if new_balance < -self._credit_limit:
            self._log_operation("withdraw", amount, "fail", self._balance)
            return False

        self._balance = new_balance
        self._log_operation("withdraw", amount, "success", self._balance)
        return True

    def get_credit_available(self) -> float:
        """
        Возвращает доступный кредитный лимит.

        :return: Доступный кредит (>= 0).
        """
        return self._credit_limit + self._balance

    def get_credit_limit(self) -> float:
        """
        Возвращает установленный кредитный лимит.

        :return: Кредитный лимит.
        """
        return self._credit_limit
```
